# Text  Classification and Hiperparameter Search

The goal of this notebook is to show how to perform a text classification and hyperparameter search using scikitlearn. The machine learning algorithms that were used ara Naive bayes (NB), Logistic Regression (LR) and KNN.

This notebook shows only how to use the param optmizer class that cam be found at find_best_hyperparameters.py

the process to immplement the text classification is the following:

1. Read the dataset
2. Instanciate the param_optmizer class
3. split the dataset int train and test set. When we do it, the train and test split is saved into a variable inside the class
4. Perform the encoding into the dataset. this is a preprocessing step that take the text file and transform it into numbers so that the algorithm can be applyed. we performed two encodings.
    - Bag of words: Used for the NB alforithm
    - TFiDF: used for the LR and KNN algorithms
5. Do a grid search: We performed a simple grid search to show how it is done. at this step, we did not focused into the params, we only selected one or two params per algorithm, however, the method suports that the user pass other params.
6. Select the best algorithm in the grid search, train it, and extract the performace metrics Acuracy and F1-score of the test dataset.

This process is done twice, once for the Dmoz-Sports.csv dataset and another for the Dmoz-COmputers.csv dataset.

We strongly support you to take a look at the code in [find_best_hyperparameters.py](https://github.com/pedrobpio/nlp/blob/main/Classification/find_best_hyperparameters.py) for more datails in the methods implementations.
    

```python
from find_best_hyperparameters import Param_optmizer
import numpy as np
import pandas as pd
```

At this exemple we are going to run the grid otimization on two datasets.

first we are using the **Dmoz-Sports.csv** and latter we are going to use the **Dmoz-Computers.csv**

# Dmoz-Sports.csv

First we need to load the dataset and instanciate the optimizer

```python
df = pd.read_csv('Data/Dmoz-Sports.csv')
X = df['text'].values
labels = df['class'].values
```

```python
optimizer = Param_optmizer()
```

Then we nee to split the dataset into train and test sets, this can be done using the train_test_split method that simply applies the scikit-learn train test split function and save it inside some classes variables

```python
optimizer.train_test_split(X, labels, test_size=0.2, random_state=42)
```

Afterwards we extracts the bag of word and the TFiDF of the datasets, this was done inside the preprocessing method

```python
optimizer.preprocessing()
```

### grid search

Now we can start the grid search. We implemented the grid search with NB, LR and KNN, which can be executed as follows:

```python
grid_NB = optimizer.NB_Grid_search()
```

```python
grid_LR = optimizer.LR_Grid_search()
```

```python
grid_KNN = optimizer.KNN_Grid_search()
```

If you want, you can also pass the params as follows:


```python
grid_KNN = optimizer.KNN_Grid_search(params = {'n_neighbors': [4, 8, 12, 16, 20, 24, 28, 32, 36]})
```

The grid search method also saves the best model inside the class:

```python
print(grid_NB.best_estimator_ == optimizer.NB)
print(grid_LR.best_estimator_ == optimizer.LR)
print(grid_KNN.best_estimator_ == optimizer.KNN)

print(optimizer.NB)
print(optimizer.LR)
print(optimizer.KNN)
```

### Saving the grid seach results
If you want, you can also saave all grid searchs results into data frames and save it into csv files:

```python
df_NB = pd.DataFrame(grid_NB.cv_results_)
df_LR = pd.DataFrame(grid_LR.cv_results_)
df_KNN = pd.DataFrame(grid_KNN.cv_results_)
```

```python
df_NB
```

```python
df_LR
```

```python
df_KNN
```

Finally we can save the csv file:

```python
df_NB.to_csv('Grid_results/grid_search_NB_Dmoz-Sports.csv')
df_LR.to_csv('Grid_results/grid_search_LR_Dmoz-Sports.csv')
df_KNN.to_csv('Grid_results/grid_search_KNN_Dmoz-Sports.csv')
```

### Retraining the best models

Now we have to retrain the model using all the train dataset and extract the models metrics. this can be done using the train_and_run_best_ methids:

```python
report, micro_f1 = optimizer.train_and_run_best_NB(NB = optimizer.NB)
print(report)
print(f'f1-micro = {micro_f1}')
```

```python
report, micro_f1 = optimizer.train_and_run_best_LR(LR = optimizer.LR)
print(report)
print(f'f1-micro = {micro_f1}')
```

```python
report, micro_f1 = optimizer.train_and_run_best_KNN(KNN = optimizer.KNN)
print(report)
print(f'f1-micro = {micro_f1}')
```
